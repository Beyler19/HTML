```python
# 1) x dəyişəni verilmişdir
x = 5
if x > 0:
    print("1) müsbət")
elif x < 0:
    print("1) mənfi")
else:
    print("1) sıfır")

# 2) n rəqəmi cütdürsə
n = 7
if n % 2 == 0:
    print("2) cüt")
else:
    print("2) tək")

# 3) a, b, c rəqəmləri verilmişdir
a, b, c = 4, 9, 2
print("3) Ən böyük:", max(a, b, c))

# 4) day dəyişəni 1-7 arası rəqəmdir
day = 5
days = {
    1: "Bazar ertəsi",
    2: "Çərşənbə axşamı",
    3: "Çərşənbə",
    4: "Cümə axşamı",
    5: "Cümə",
    6: "Şənbə",
    7: "Bazar"
}
print("4)", days.get(day, "Yanlış gün"))

# 5) temp dəyişəni temperaturdur
temp = 18
if temp < 0:
    print("5) soyuq")
elif temp <= 20:
    print("5) normal")
else:
    print("5) isti")

# 6) password adlı string verilmişdir
password = "sifre123"
if len(password) < 8:
    print("6) qısa")
elif len(password) <= 12:
    print("6) orta")
else:
    print("6) uzun")

# 7) x rəqəmi həm 3-ə, həm də 5-ə bölünürsə
x = 15
if x % 3 == 0 and x % 5 == 0:
    print("7) 3 və 5")
elif x % 3 == 0:
    print("7) 3")
elif x % 5 == 0:
    print("7) 5")
else:
    print("7) heç biri")

# 8) 0-dan 20-yə qədər cüt rəqəmləri çap etsin
print("8) 0-20 arası cütlər:")
for i in range(0, 21, 2):
    print(i, end=' ')
print()

# 9) "Bağda ərik var idi …" stringinin hər elementini for ilə çap edin
text = "Bağda ərik var idi …"
print("9) hər simvol:")
for ch in text:
    print(ch)

# 10) 1-dən 10-a qədər rəqəmləri çap edin, amma 3 xaric
print("10) 3 xaric:")
for i in range(1, 11):
    if i == 3:
        continue
    print(i, end=' ')
print()

# 11) 1-dən başlayaraq ilk 5-ə bölünən rəqəmi tapın
i = 1
print("11) İlk 5-ə bölünən:")
while True:
    if i % 5 == 0:
        print(i)
        break
    i += 1

# 12) (1, 3, 5, 7, 9) listində 5-i tap və indeksini çap et
numbers = [1, 3, 5, 7, 9]
print("12) 5-in indeksi:")
for idx, num in enumerate(numbers):
    if num == 5:
        print(idx)
        break
```

```python
# 1) salam adlı funksiya
def salam():
    print("1) Salam, Dünya!")

salam()

# 2) kub_hesabla adlı funksiya
def kub_hesabla(n):
    return n ** 3

print("2)", kub_hesabla(3))

# 3) birlesdir adlı funksiya
def birlesdir(soz1, soz2):
    return soz1 + " " + soz2

print("3)", birlesdir("Salam", "Dünya"))

# 4) cap_et adlı funksiya
def cap_et(lst):
    print("4) List elementləri:")
    for element in lst:
        print(element)

cap_et(["alma", "armud", "nar"])

# 5) toplam adlı funksiya
def toplam(*args):
    return sum(args)

print("5)", toplam(1, 2, 3, 4, 5))

# 6) ortalama adlı funksiya
def ortalama(*args):
    if len(args) == 0:
        return "Rəqəm yoxdur"
    return sum(args) / len(args)

print("6)", ortalama(10, 20, 30))
print("6-2)", ortalama())

# 7) adlar_rəqəmlər adlı funksiya
def adlar_rəqəmlər(**kwargs):
    print("7) Ad və rəqəmlər:")
    for ad, rəqəm in kwargs.items():
        print(f"ad: {ad}, rəqəm: {rəqəm}")

adlar_rəqəmlər(bir=1, iki=2, uc=3)

# 8) tip_yoxla adlı funksiya
def tip_yoxla(dəyər):
    if isinstance(dəyər, str):
        print("8) mətn")
    elif isinstance(dəyər, (int, float)):
        print(" rəqəm")
    else:
        print(" başqa")

tip_yoxla("salam")
tip_yoxla(123)
tip_yoxla([1, 2, 3])

# 9) yas_kateqoriya adlı funksiya
def yas_kateqoriya():
    try:
        yas = int(input("9) Yaşınızı daxil edin: "))
        if yas < 18:
            print("Gənc")
        else:
            print("Yetkin")
    except ValueError:
        print("Düzgün yaş daxil edin!")

# Funksiyanı çağırırıq:
yas_kateqoriya()


# 10) soz_uzunluq adlı funksiya
def soz_uzunluq():
    soz = input("10) Bir söz daxil edin: ")
    print("Söz uzunluğu:", len(soz))

# Funksiyanı çağırırıq:
soz_uzunluq()

```
